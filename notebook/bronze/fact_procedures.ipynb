```python
path=f"/Volumes/hdfc_ergo/dummy_data/raw/Fact_Procedures.csv"
df = spark.read.format("csv").options(header=True,inferschema=True).load(path)
```

```python
from pyspark.sql.functions import current_timestamp, col
df = df.withColumn("_inserted_at", current_timestamp())\
        .withColumn("_source_file", col("_metadata.file_path"))
new_col = [c.lower().replace(" ","_") for c in df.columns]
df = df.toDF(*new_col)
display(df.limit(5))
```

```python
df.write.mode("overwrite").saveAsTable("hdfc_ergo.hdfc_bronze.fact_procedures")
```
