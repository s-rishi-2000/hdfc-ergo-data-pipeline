```python
path = f"/Volumes/hdfc_ergo/dummy_data/raw/Dim_CPT_Codes.csv"
df = spark.read.format("csv").options(header=True, inferschema=False).load(path)
display(df.limit(5))
```

```python
from pyspark.sql.functions import current_timestamp, col

# Add metadata columns first, before renaming
df = df.withColumn("_ingested_at", current_timestamp()) \
       .withColumn("_source_file", col("_metadata.file_path"))

# Now rename all columns (including the new ones)
new_col = [c.lower().replace(" ","_").replace("-"," ") for c in df.columns]
df = df.toDF(*new_col)

display(df.limit(5))
```

```python
df.write.mode("overwrite").saveAsTable("hdfc_ergo.hdfc_bronze.dim_cpt_codes")
display(df.limit(5))
```
