```python
df_silver = spark.read.table("he_prod_incen_analytics_dbw_01.silver.dim_plan_types")
```

```python
from pyspark.sql.functions import row_number
from pyspark.sql.window import Window

window_spec = Window.orderBy("plan_type_id")
df_gold = df_silver.withColumn("plan_sk", row_number().over(window_spec))

df_gold = df_gold.select(
    "plan_sk", "plan_type_id", "plan_type_name", "sum_insured_min", "sum_insured_max",
    "copay_percentage_min", "copay_percentage_max", "room_rent_limit", "network_requirement",
    "pre_hospitalization_days", "post_hospitalization_days", "maternity_coverage", "critical_illness_covered"
)

display(df_gold.limit(5))
```

```python
df_gold.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("he_prod_incen_analytics_dbw_01.gold.dim_plan_types")
print("✅ Successfully wrote Dim_Plan_Types to Gold layer.")
```
