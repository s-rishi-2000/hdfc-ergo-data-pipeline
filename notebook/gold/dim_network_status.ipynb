```python
df_silver = spark.read.table("he_prod_incen_analytics_dbw_01.silver.dim_network_status")
```

```python
from pyspark.sql.functions import row_number
from pyspark.sql.window import Window

window_spec = Window.orderBy("network_status_id")
df_gold = df_silver.withColumn("network_status_sk", row_number().over(window_spec))

df_gold = df_gold.select(
    "network_status_sk", "network_status_id", "network_status_name", 
    "coinsurance_percentage", "deductible_applied", "prior_auth_required"
)

display(df_gold.limit(5))
```

```python
df_gold.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("he_prod_incen_analytics_dbw_01.gold.dim_network_status")
print("✅ Successfully wrote Dim_Network_Status to Gold layer.")
```
