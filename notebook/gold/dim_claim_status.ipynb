```python
df_silver = spark.read.table("he_prod_incen_analytics_dbw_01.silver.dim_claim_status")
```

```python
from pyspark.sql.functions import row_number
from pyspark.sql.window import Window

window_spec = Window.orderBy("claim_status_id")
df_gold = df_silver.withColumn("claim_status_sk", row_number().over(window_spec))

df_gold = df_gold.select(
    "claim_status_sk", "claim_status_id", "claim_status_name", "status_description"
)

display(df_gold.limit(5))
```

```python
df_gold.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("he_prod_incen_analytics_dbw_01.gold.dim_claim_status")
print("✅ Successfully wrote Dim_Claim_Status to Gold layer.")
```
